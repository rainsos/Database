[문제8]

SQL 언어에 해당하지 않는 것을 골라주세요

- 데이터 처리 언어(DML)
- 데이터 제어 언어(DCL)
- 데이터 연결 언어(DCL)
- 데이터 정의 언어(DDL)

정답 : 데이터 연결 언어(DCL)

- SQL 언어의 공식 분류
1.DML (Data Manipulation Language)
데이터 처리 언어로, 데이터 조회, 삽입, 수정, 삭제 등을 수행한다.
예시: SELECT, INSERT, UPDATE, DELETE

2.DDL (Data Definition Language)
데이터 정의 언어로, 데이터베이스 구조를 정의하는 명령어입니다. 테이블, 뷰, 인덱스 등의 구성을 정의한다.
예시: CREATE, ALTER, DROP

3.DCL (Data Control Language)
데이터 제어 언어로, 사용자 권한을 관리하는 명령어.
예시: GRANT, REVOKE

4.TCL (Transaction Control Language)
트랜잭션 제어 언어로, 데이터베이스 내의 트랜잭션 처리와 관련된 명령어. 트랜잭션을 시작, 커밋, 롤백하는 등의 작업을 다룬다.
